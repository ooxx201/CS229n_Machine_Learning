Since $\nabla_{A^{T}} trABA^{T}C = B^{T}A^{T}C^{C} + BA^{T}C$ and $A^{T} = A$
\begin{equation}
\begin{split}
\nabla_{x} \frac{1}{2}x^{T}Ax &= \nabla_{x} tr\frac{1}{2}x^{T}Ax\\
                              &= \nabla_{x} tr\frac{1}{2}(x^{T}) (A) (x^{T})^{T}I\\
                              &= \frac{1}{2} (A^{T}(x^{T})^{T}I^{T} + A(x^{T})^{T}I)\\
                              &= Ax
\end{split}
\end{equation}
Since $\nabla_{A} trAB = B^{T}$ and $trAB = trBA$ 
\begin{equation}
\begin{split}
\nabla_{x} b^{T}x &= \nabla_{x} trb^{T}x\\
                  &= \nabla_{x} trxb^{T}\\
                  &= b
\end{split}
\end{equation}
Thus:  
\begin{equation}
\begin{split}
\nabla_{x} f(x) &= \nabla_{x} \frac{1}{2}x^{T}Ax + \nabla_{x} b^{T}x\\
                &= Ax + b
\end{split}
\end{equation}

#### (a)
Since $A = T \Lambda T^{-1}$, we have $AT = T\Lambda$, thus 
\begin{equation}
\begin{split}
A\begin{bmatrix}t^{(1)}&\dots&t^{(n)} \end{bmatrix} &= \begin{bmatrix}t^{(1)}&\dots&t^{(n)} \end{bmatrix} \Lambda \\
                                                    &= \begin{bmatrix}\lambda_{1}t^{(1)}&\dots&\lambda_{n}t^{(n)}\end{bmatrix} \\
\end{split}
\end{equation}
Thus $At^{(i)} = \lambda_{i} t^{(i)}$.  
$(t^{(i)}, \lambda_{i})$ is eigenvalues/eigenvector pair.
